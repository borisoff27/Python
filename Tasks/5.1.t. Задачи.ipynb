# Исправь ошибки

Программа запрашивает у пользователя его настроение и печатает рекомендацию товара:

1) Если настроение "весёлое", то "Попробуйте апельсиновый лимонад!"
2) Если настроение "спокойное", то "Как насчёт ягодного чая?"
3) Во всех остальных случаях - "Вам понравится вишнёвый йогурт!"

```python
mood = input('Ваше настроение:')
if mood = весёлое:
    print('Попробуйте апельсиновый лимонад!')
elif mood = спокойное:
    print('Как насчёт ягодного чая?')
elif:
    print('Вам понравится вишнёвый йогурт!')
```
